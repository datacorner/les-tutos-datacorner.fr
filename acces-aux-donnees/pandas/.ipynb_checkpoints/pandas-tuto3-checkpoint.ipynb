### Ajout d'une colonne de valeur nulle

```python
tabzero = []
for i in range(0, train.shape[0]):
    tabzero.append(0)
Cab_T = pd.Series(tabzero, name="Ma Colonne")
train = train.join(Cab_T)
train.head(3)
```
